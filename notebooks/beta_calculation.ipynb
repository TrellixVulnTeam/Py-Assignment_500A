```python
import pandas as pd
import numpy as np
import datetime as dt
df = pd.read_excel('data/task2_stock_data.xlsx')
df
```

```python
from scipy.stats import mstats

def using_mstats_df(df):
    return df.apply(using_mstats, axis=0)

def using_mstats(s):
    return mstats.winsorize(s, limits=[0.05, 0.05])

df.set_index('date', inplace=True)
grouped = df.groupby(level='date')
df = grouped.apply(using_mstats_df)
df.reset_index(inplace=True)
```

```python
AS_OF_DATE = pd.to_datetime("2021-10-31")
WINDOW = "1y"
FREQUENCY = "bi-weekly"
# FREQUENCY = "quarterly"
today = str(dt.date.today())
year_ago = (AS_OF_DATE - pd.DateOffset(years=int(WINDOW[0]))).to_pydatetime()

frequency_dict = {"daily": "D", "weekly": "W", "yearly": "Y", "monthly": "M", 'bi-weekly': 'SM', 'quarterly': 'Q'}
print("-- Frequency --> ", "%s [%s]" % (FREQUENCY, frequency_dict[FREQUENCY]))
print("-- Date Range -->", "From: %s, To: %s" % (AS_OF_DATE, year_ago))
input_df = df[(df['date'] >= year_ago) & (df['date'] < AS_OF_DATE)]
input_df
```

```python
input_df.reset_index(inplace=True)
input_df.set_index('date', inplace=True)
log_returns = np.log(input_df/input_df.shift())
log_returns
```

```python
log_returns.reset_index(inplace=True)
```

```python
lr = log_returns.resample(frequency_dict[FREQUENCY], on='date').mean()
```

```python
covariance = lr.cov()
covariance
```

```python
variance = lr["SPY US Equity"].var()
variance
```

```python
beta = covariance.loc["APTMU US Equity", "SPY US Equity"]/variance
beta
```

```python
beta_df = covariance.loc["SPY US Equity"]/variance
beta_df
```
